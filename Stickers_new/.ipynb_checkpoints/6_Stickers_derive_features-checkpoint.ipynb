```python
#consecutive days computation
from pyspark.sql import functions as F

new_base = active_base.groupby("user_id_n").agg(F.collect_list("date").alias("date_list"))
```

```python
from pyspark.sql.types import FloatType
from datetime import datetime,timedelta
#function to derive average consecutive days
def con_days(raw):
    raw = list(set(raw))
    n = len(raw)
    raw = sorted(raw, key=lambda d: map(int, d.split('-')))
    no_of_days = 0
    no_of_groups = 0
    group = 0
    total_days = 0
    res = []
    for i in range(0,n-1):
        current_date = datetime.strptime(raw[i], "%Y-%m-%d")
        current_date_mod = datetime.strftime(current_date, "%Y-%m-%d")
        con_date = current_date + timedelta(days=1)
        con_date_mod = datetime.strftime(con_date, "%Y-%m-%d")

        next_date = datetime.strptime(raw[i+1], "%Y-%m-%d")
        next_date_mod = datetime.strftime(next_date, "%Y-%m-%d")
        if con_date_mod == next_date_mod :
            no_of_days += 1
            total_days += 1
            if no_of_days == 1:
                group += 1
                no_of_groups = group

        else:
            no_of_days = 0
    res.append(total_days)
    return total_days

consecutive_days = udf(con_days,StringType())

def con_groups(raw):
    raw = list(set(raw))
    n = len(raw)
    raw = sorted(raw, key=lambda d: map(int, d.split('-')))
    no_of_days = 0
    no_of_groups = 0
    group = 0
    total_days = 0
    res = []
    for i in range(0,n-1):
        current_date = datetime.strptime(raw[i], "%Y-%m-%d")
        current_date_mod = datetime.strftime(current_date, "%Y-%m-%d")
        con_date = current_date + timedelta(days=1)
        con_date_mod = datetime.strftime(con_date, "%Y-%m-%d")

        next_date = datetime.strptime(raw[i+1], "%Y-%m-%d")
        next_date_mod = datetime.strftime(next_date, "%Y-%m-%d")
        if con_date_mod == next_date_mod :
            no_of_days += 1
            total_days += 1
            if no_of_days == 1:
                group += 1
                no_of_groups = group

        else:
            no_of_days = 0
    return no_of_groups

groups = udf(con_groups,StringType())
```

```python
new_base = new_base.withColumn("Consective_days",consecutive_days('date_list'))
new_base = new_base.withColumn("Consective_groups",groups('date_list'))
new_base = new_base.withColumn("avg_con_days",when((col('Consective_days') != 0)  & (col('Consective_groups') != 0) , round((col('Consective_days')/col('Consective_groups')))).otherwise(0))
```

```python
#average consecutive distribution
data = new_base.groupby('avg_con_days').agg(countDistinct('user_id_n').alias('cnt'))
total = data.select("cnt").agg({"cnt": "sum"}).collect().pop()['sum(cnt)']
data = data.withColumn("percentage", (data['cnt']/total) * 100)
data.sort(col('avg_con_days').desc()).show(data.count())
```

```python
final_data = final_data.withColumn('filter_1',when(col('numofdays')<7,0.25).otherwise(when((col('numofdays') >=7 ) & (col('numofdays') < 15 ),0.5).otherwise(when(col('numofdays') >=15 ,1).otherwise('null'))))
final_data = final_data.withColumn('filter_2',when((col('avg_con_days') >=1 ) & (col('avg_con_days') < 7 ),0.25).otherwise(when((col('avg_con_days') >=7 ) & (col('avg_con_days') < 15 ),0.5).otherwise(when(col('avg_con_days') >=15 ,1).otherwise('null'))))
final_data = final_data.withColumn('filter_3',when((col('sticker_packs_sent') >= 1 ) & (col('sticker_packs_sent') < 100 ),0.25).otherwise(when((col('sticker_packs_sent') >=100 ) & (col('sticker_packs_sent') < 1000 ),0.5).otherwise(when(col('sticker_packs_sent') >=1000 ,1).otherwise('null'))))
```

```python
new = final_data.where(col('filter_1') != 'null').where(col('filter_2') != 'null').where(col('filter_2') != 'null')
```

```python
new = new.withColumn("value",col('filter_1')+col('filter_2')+col('filter_3'))
```
